EXP 5 bloom Filter

```python
import hashlib
import math

class BloomFilter:
    def __init__(self, size, hash_count):
        self.size = size
        self.hash_count = hash_count
        self.bit_array = [0] * size

    def _hashes(self, item):
        """Generate a list of hash values for the item using different hash functions."""
        hashes = []
        for i in range(self.hash_count):
            hash_result = int(hashlib.sha256(item.encode('utf-8')).hexdigest(), 16) + i
            hashes.append(hash_result % self.size)
        return hashes

    def add(self, item):
        """Add an item to the Bloom filter."""
        for index in self._hashes(item):
            self.bit_array[index] = 1

    def check(self, item):
        """Check if an item is possibly in the Bloom filter."""
        for index in self._hashes(item):
            if self.bit_array[index] == 0:
                return False
        return True

# Example usage:
size = 1000  # size of the bit array
hash_count = 5  # number of hash functions

bloom = BloomFilter(size, hash_count)

# Add some items
bloom.add("apple")
bloom.add("banana")
bloom.add("cherry")

# Check for existence
print(bloom.check("apple"))   # Output: True
print(bloom.check("banana"))  # Output: True
print(bloom.check("grape"))   # Output: False
print(bloom.check("cherry"))  # Output: True


```

EXP 6 K-means

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate some random data for clustering
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualize the data
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title('Generated Data')
plt.show()

# Apply K-Means algorithm
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

# Get cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Plot the data points with the cluster centers
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering')
plt.show()

```

Exp 7 Page rank

```python
from collections import defaultdict

def pagerank_mapper(page_id, page_rank, outbound_links):
    """Emit contributions to linked pages and pass the list of outbound links."""
    contributions = []
    if outbound_links:
        num_links = len(outbound_links)
        for link in outbound_links:
            contributions.append((link, page_rank / num_links))
    # Emit the page's outbound links as well to reconstruct the graph in the reducer
    contributions.append((page_id, outbound_links))
    return contributions

def pagerank_reducer(page_id, incoming_info):
    """Aggregate the contributions and calculate the new rank."""
    damping_factor = 0.85
    base_rank = (1.0 - damping_factor)

    new_rank = base_rank
    outbound_links = []

    for info in incoming_info:
        if isinstance(info, list):  # This is the list of outbound links
            outbound_links = info
        else:  # This is a contribution to the rank
            new_rank += damping_factor * info

    return page_id, new_rank, outbound_links

def run_pagerank(pages, num_iterations):
    """Run the PageRank algorithm for a specified number of iterations."""
    ranks = {page: (1.0, outbound_links) for page, outbound_links in pages.items()}  # Initialize rank for each page

    for i in range(num_iterations):
        mapper_outputs = defaultdict(list)

        # Mapper phase
        for page_id, (rank, outbound_links) in ranks.items():
            contributions = pagerank_mapper(page_id, rank, outbound_links)
            for link, contribution in contributions:
                mapper_outputs[link].append(contribution)

        # Reducer phase
        new_ranks = {}
        for page_id, incoming_info in mapper_outputs.items():
            new_page_id, new_rank, outbound_links = pagerank_reducer(page_id, incoming_info)
            new_ranks[new_page_id] = (new_rank, outbound_links)

        ranks = new_ranks

    return ranks

# Example graph
pages = {
    "A": ["B", "C"],
    "B": ["C"],
    "C": ["A"],
    "D": ["C"],
}

# Run PageRank
num_iterations = 10
final_ranks = run_pagerank(pages, num_iterations)

# Print final ranks
for page, (rank, _) in final_ranks.items():
    print(f"Page {page} has rank: {rank:.4f}")

```
